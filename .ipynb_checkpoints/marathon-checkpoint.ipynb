```python
import pandas as pd
import seaborn as sns
```

# Importing the data

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.head()
```

```python
df.dtypes
```

# Cleaning up the data

```python
# get race data from only USA with 50km or 50 miles in 2020
```

```python
df[df['Event distance/length'] == '50mi']
```

```python
df[df['Event distance/length'] == '50km']
```

```python
df[df['Event distance/length'].isin(['50km', '50mi'])]
```

```python
#find events for 2020 only
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]
```

```python
#df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']

#OR

df[df['Event name'].str.contains(r'\(USA\)$')]
```

```python
#combine all the conditions
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.contains(r'\(USA\)$'))]
```

```python
df2.head(10)
```

```python
#remove (USA) from event name
df2['Event name'] = df2['Event name'].str.replace('(USA)', '')
df2
```

```python
df2[df2['Event name'].str.contains('(USA)')]
```

```python
#remove the random USA in the df

df2 = df2[~df2['Event name'].str.contains('USA')]
df2
```

```python
#clean up athelet age
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
df2
```

```python
# remove h from athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.replace(' h', '')
df2
```

```python
#drop columns: Athlete club, Athlete country, Athlete year of birth, Athlete age category
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
df2
```

```python
#clean up null values
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#checking for duplicates
df2[df2.duplicated() == True]
```

```python
#reset index
df2.reset_index(drop = True)
```

```python
# fix types
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.head()
```

```python
#rename columns
df2 = df2.rename(columns = { 'Year of event' : 'year',
                             'Event dates' : 'race_day',
                             'Event name' : 'race_name',
                             'Event distance/length' :  'race_length' ,
                             'Event number of finishers' : 'race_number_of_finishers',
                             'Athlete performance' : 'athlete_performance',
                             'Athlete gender' : 'athlete_gender',
                             'Athlete average speed' : 'athlete_average_speed',
                             'Athlete ID' : 'athlete_id'  
})
```

```python
df2.head()
```

```python
#reorder columns
df3 = df2[['race_day','race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
df3
```

# Questions

```python
# 1. Difference in speed for the 50k, 50mi in terms of male vs female
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# 2. What age groups are the best in the 50m race (20+ races minimum)
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19').head(15)
```

```python
# 2. What age groups are the worst in the 50m race (20+ races minimum)
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>19').head(15)
```

```python
#Seasons for the data -> Slower in summer than winter?

#spring = 3-5
#summer = 6-8
#fall = 9-11
#winter = 12-2
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
df3.head()
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.head()
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```

```python
#50 miles only
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False)
```
